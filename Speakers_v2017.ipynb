```python
import pandas as pd
import numpy as np
import time
from os import listdir
```

```python
df2 = pd.read_parquet('/content/drive/MyDrive/ADA/Project datasets/speaker_attributes.parquet')
```

```python
df2.sample(2)
```

```python
speakers_all = pd.read_csv('/content/drive/MyDrive/ADA/ada-sports-quotes/sport-quotes-2017.csv.bz2', compression='bz2', nrows=1)
```

```python
for chunk in pd.read_csv('/content/drive/MyDrive/ADA/ada-sports-quotes/sport-quotes-2017.csv.bz2', compression='bz2', chunksize=10**5):
      chunk['qids'] = chunk['qids'].str.strip('[]')
      chunk['qids'] = chunk['qids'].str.strip('\'\'')
      speakers=df2[df2['id'].isin(chunk['qids'])]
      speakers_all = pd.concat([speakers_all, speakers])
```

```python
speakers_all.drop(labels=0, inplace=True)
speakers_all.shape
```

```python
politicians=speakers_all[~(speakers_all['party'].isna())]
politicians.shape
```

```python
df3 = pd.read_csv('/content/drive/MyDrive/ADA/Project datasets/wikidata_labels_descriptions_quotebank.csv.bz2', compression='bz2', index_col='QID')
```

```python
partys=df3[df3.index.isin((politicians['party'].str.get(0)))]
partys.sample(10)
```
